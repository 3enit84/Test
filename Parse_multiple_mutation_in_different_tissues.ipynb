```python
#Input file should be sorted by mutation (second column)

file = open('All_ndividual_mutation_count.tsv','r')
newfile = open('All_ndividual_mutation_coun_out.tsv','w')
line = file.readline()#read the title line
OrganLst = ['Breast','Ovary','Prostate','Lung','Colon','Pancreas']#list of organs
for line in file: 
    Mut = str.split(line,'\t')[1]#set first mutation in the file
    Organ = str.split(line,'\t')[0]
    Count = int(str.split(line,'\t')[2])
    break#stop after reading one line
OrganNdx = [0,0,0,0,0,0]#list of indecies
for line in file:
    if Mut == str.split(line,'\t')[1]:
        Organ = str.split(line,'\t')[0]
        Count = int(str.split(line,'\t')[2])
        OrganNdx[OrganLst.index(Organ)] = Count
    else:
        OrganNdx[OrganLst.index(Organ)] = Count
        newfile.write(Mut)
        NumOrg = SumMut = 0#number of organs with a mutation, total number of mutations
        for i in range(len(OrganNdx)):
            if OrganNdx[i] != 0:
                NumOrg += 1
                SumMut += OrganNdx[i]
            newfile.write('\t'+str(OrganNdx[i]))
        newfile.write('\t'+str(SumMut)+'\t'+str(NumOrg)+'\n')
        ############reset everything############
        OrganNdx = [0,0,0,0,0,0]
        Mut = str.split(line,'\t')[1]
        Organ = str.split(line,'\t')[0]
        Count = int(str.split(line,'\t')[2])
        OrganNdx[OrganLst.index(Organ)] = Count
newfile.close()
```
