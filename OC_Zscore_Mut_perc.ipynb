#### count significant Z-scores per gene

```python
import os, glob

newfile = open('OC_Z_cntd.tsv', 'w')

FileLst = []
newfile.write('Gene')
for filename in glob.glob("Ovary_Zscore*"):
    print(filename)
    FileLst = FileLst + [open(filename, "r")]#open all Zscore files  
    newfile.write('\t'+filename[:-4])#make a title
newfile.write('\n')
for file in FileLst:
    for i in range(2):#skip first 2 lines in each file
        line = file.readline()
rec = str.split(line,'\t')
while len(rec) > 99:#experiments should have more than 100 patients
    check = 0
    for file in FileLst:
        for line in file:
            rec = str.split(line,'\t')
            if len(rec) <100:
                print('break')
                break
            Zcnt = 0
            rec = str.split(line,'\t')
            if check == 0:
                newfile.write(rec[0])#write gene symbol only once
                check = 1
            for i in range(1,len(rec)):#first entry is a gene symbol
                if str.strip(rec[i]) != 'NA':
                    if float(str.strip(rec[i])) > 1.96 or float(str.strip(rec[i])) < -1.96:
                        if rec[0] == 'OR7D2':
                            print(file,rec[i])
                        Zcnt += 1
            
            newfile.write('\t'+str(Zcnt*100.0/(len(rec)-1)))
            #print(file,rec[0],Zcnt)
            break #read only one line
    print(len(rec))
    print(rec[0])
    newfile.write('\n')
    if rec[0] == 'TAAR9':
        print('break')
        break

newfile.close()
    
```
